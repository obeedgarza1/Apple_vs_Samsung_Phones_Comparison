```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

First file corrresponding to the Apple phones

```python
df1 = pd.read_csv(r"C:\Users\PC\Downloads\Iphone 1.csv")
df1.head()
```

Second file corrresponding to the Samsung phones

```python
df2 = pd.read_csv(r"C:\Users\PC\Downloads\Samsung 1.csv")
df2.head()
```

Combining both into the df

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
df.head()
```

Creating a brand column to having it as the first index, then the Phone Model

```python
df['Brand'] = df['Phone Model'].apply(lambda x: 'Apple' if 'iPhone' in x else 'Samsung')
df.set_index(['Brand','Phone Model'],inplace=True)
df.sort_index().head()
```

```python
df.drop(columns='Additional Features',inplace=True)
df.head()
```

Splitting the Estimated Price Range into 2 columns and cleaning the data of this 2 columns

```python
df[['Price 1','Price 2']] = df['Estimated Price Range (USD)'].str.split('-',expand=True)
df['Price 1'] = df['Price 1'].str.replace('$', '')
df['Price 2'] = df['Price 2'].str.replace('$', '')
df['Price 2'] = df['Price 2'].str.replace(' ', '')
df['Price 2'] = df['Price 2'].str.replace(',', '')
df.head()
```

Creating the column 'Average Price' by calculating the mean of Price 1 and Price 2, then droping this 2 columns

```python
df['Average Price'] = df[['Price 1','Price 2']].astype(float).mean(axis=1)
df.drop(columns=['Estimated Price Range (USD)','Price 1','Price 2'],inplace=True)
df.head()
```

Splitting the column 'Accessories' into various columns getting a 'Earphones' and 'Power adapter' column

```python
df[['Earphones','Power adapter','Extra 1','Extra 2', 'Extra 3']] = df['Accessories'].str.split(',',expand=True)
df.head()
```

Replacing some information from a column to another, where this information should be, then dropping the Extra columns

```python
string = df['Earphones'].str.contains('Charger')
df.loc[string, 'Extra 1'] = df.loc[string, 'Earphones']
df.loc[string, 'Earphones'] = df.loc[string, 'Earphones'].str.replace('Charger','')
```

```python
string = df['Power adapter'].astype(str).str.contains('Earphones')
df.loc[string, 'Earphones'] = df.loc[string, 'Power adapter']
df.loc[string, 'Power adapter'] = df.loc[string, 'Power adapter'].astype(str).str.replace('Earphones','')
```

```python
string = df['Power adapter'].astype(str).str.contains('EarPods')
df.loc[string, 'Earphones'] = df.loc[string, 'Power adapter']
df.loc[string, 'Power adapter'] = df.loc[string, 'Power adapter'].astype(str).str.replace('EarPods','')
```

```python
string = df['Power adapter'].astype(str).str.contains('Headphones')
df.loc[string, 'Earphones'] = df.loc[string, 'Power adapter']
df.loc[string, 'Power adapter'] = df.loc[string, 'Power adapter'].astype(str).str.replace('Headphones','')
```

```python
string = df['Extra 1'].astype(str).str.contains('Charger')
df.loc[string, 'Power adapter'] = df.loc[string,'Extra 1']
df.loc[string,'Extra 1'] = df.loc[string,'Extra 1'].astype(str).str.replace('Charger','')
```

```python
string = df['Extra 1'].astype(str).str.contains('adapter')
df.loc[string, 'Power adapter'] = df.loc[string,'Extra 1']
df.loc[string,'Extra 1'] = df.loc[string,'Extra 1'].astype(str).str.replace('adapter','')
```

```python
df.drop(columns='Extra 1',inplace=True)
df.drop(columns='Extra 2',inplace=True)
df.drop(columns='Extra 3',inplace=True)
df.head()
```

```python
df.loc['Apple']
```

Adding information to the Power adapter column if the phone if fact has a adapter

```python
df.loc[('Apple',['iPhone (Original)','iPhone 3G','iPhone 4','iPhone 6','iPhone 7','iPhone SE (1st gen)','iPhone X','iPhone 11',]),'Power adapter'] = 'Power Adapter'
df.loc['Apple']
```

```python
df.loc[(df['Earphones'].str.contains('No')), 'Earphones']=''
df.loc[(df['Power adapter'].str.contains('No')), 'Power adapter']=''
df.head()
```

Chaning the columns of 'Earphones' and 'Power adapter' if the Phone 'Model' has this acceories then 1, if not 0, then dropping the column that was not necessary anymore

```python
df['Earphones'] = df['Earphones'].fillna('').apply(lambda x: 0 if x == '' else 1)
df['Power adapter'] = df['Power adapter'].fillna('').apply(lambda x: 0 if x == '' else 1)
df
```

```python
df.drop(columns='Accessories',inplace=True)
```

```python
df.to_csv('DF Phones 1.csv')
```

Next part is to build an Area chart

```python
apple_sales = df
samsung_sales = df
```

```python
apple_sales = apple_sales.drop(('Apple','iPhone (Original)'))
apple_sales = apple_sales.drop(('Apple','iPhone 3G'))
apple_sales = apple_sales.drop(('Apple','iPhone 3GS'))
apple_sales = apple_sales.drop(('Apple','iPhone XS'))
apple_sales = apple_sales.drop(('Apple','iPhone SE (1st gen)'))
apple_sales = apple_sales.drop(('Apple','iPhone SE (2nd Gen)'))
```

```python
apple_sales
```

```python
apple_sales = apple_sales.loc['Apple'][['Year Released','Total Sales (millions)']]
apple_sales
```

```python
samsung_sales = samsung_sales.loc['Samsung'][['Year Released','Total Sales (millions)']]
samsung_sales
```

```python
plt.fill_between(apple_sales['Year Released'],apple_sales['Total Sales (millions)'])
plt.fill_between(samsung_sales['Year Released'],samsung_sales['Total Sales (millions)'])
plt.xlabel('Year Released')
plt.ylabel('Total Sales (millions)')
plt.title('Apple and Samsung Sales Comparison')
plt.legend()
plt.show()
```
